# 3.2 Continuity and Differentiation

- **Limits and Continuity:**
  - A limit describes the behavior of a function as its input approaches a specific value.
  - A function is continuous at a point if the limit as x approaches that point equals the function value at that point.
  - Continuity ensures there are no abrupt changes in the function, which is essential for defining derivatives.

- **Differentiation:**
  - The derivative measures how a function's output changes in response to a small change in input.
  - It is defined as the limit of the difference quotient: [f(x+h) - f(x)] / h, as h approaches 0.
  - The derivative provides the best linear approximation of the function at a given point.
  - Fundamental properties include linearity and the chain rule, which are important for optimization and modeling.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin(x)

def df_exact(x):
    return np.cos(x)

def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

x_vals = np.linspace(0, 2*np.pi, 100)

y_vals = f(x_vals)
dy_exact = df_exact(x_vals)
dy_numerical = numerical_derivative(f, x_vals)

plt.figure(figsize=(10,6))
plt.plot(x_vals, y_vals, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals, dy_exact, label='Exact derivative: cos(x)', color='green', linestyle='--')
plt.plot(x_vals, dy_numerical, label='Numerical derivative', color='red', linestyle=':')
plt.xlabel('x')
plt.ylabel('Value')
plt.title('Continuity and Differentiation')
plt.legend()
plt.grid(True)
plt.show()
```

# 3.3 Taylor Theorem

- **Taylor's Theorem Overview:**
  - Taylor's theorem provides a way to approximate a differentiable function by a polynomial around a chosen point.
  - The idea is that, near a point a, the function f(x) can be approximated by f(a) plus terms involving its derivatives at a.
  - A first-order (linear) approximation is: f(x)  f(a) + f'(a)*(x - a)
  - A second-order (quadratic) approximation adds: + 0.5 * f''(a)*(x - a)^2, which gives a better approximation near a.
  - The remainder term describes the error of the approximation and becomes small when x is close to a.



```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(x)

def taylor_poly(x, a=0):
    return f(a) + f(a) * (x - a) + 0.5 * f(a) * (x - a)**2

x_vals = np.linspace(-1, 1, 100)
f_vals = f(x_vals)
taylor_vals = taylor_poly(x_vals, a=0)

plt.figure(figsize=(8,6))
plt.plot(x_vals, f_vals, label='f(x) = exp(x)', color='blue')
plt.plot(x_vals, taylor_vals, label='Quadratic Taylor Approximation', color='red', linestyle='--')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Taylor Theorem: Approximating exp(x) near x = 0')
plt.legend()
plt.grid(True)
plt.show()
```

# 3.3 Optimization with Gradient Descent

- **Gradient Descent Overview:**
  - An iterative optimization algorithm for minimizing a differentiable function.
  - At each step, the algorithm updates the current point by moving in the direction opposite to the gradient.
  - The step size (learning rate) controls how far the update moves along the negative gradient.
  - The process repeats until the function value converges to a minimum.
  - Widely used in machine learning and numerical optimization problems.


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x-2)**2 + 1

def grad_f(x):
    return 2*(x-2)

x0 = -5
alpha = 0.1
iterations = 50
x_vals = [x0]
for i in range(iterations):
    x_new = x_vals[-1] - alpha * grad_f(x_vals[-1])
    x_vals.append(x_new)

x_vals = np.array(x_vals)
x_range = np.linspace(-6, 6, 400)
plt.figure(figsize=(10,6))
plt.plot(x_range, f(x_range), label='f(x)')
plt.scatter(x_vals, f(x_vals), color='red', zorder=5)
plt.plot(x_vals, f(x_vals), color='red', linestyle='--', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Optimization with Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
